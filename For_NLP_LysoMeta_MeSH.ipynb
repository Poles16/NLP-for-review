```python
import pandas as pd

LysoMeta_MeSH=pd.read_csv("Z:\\YiboZhao\\NLP\\Esearch Data\\lysosome_AND_metabolism.txt", sep = "\t", header = None)
LysoMeta_MeSH.columns=['AbstractText', 'PMID', 'DescriptorName', 'Year']
```

```python
from bunkatech import Bunka

# Instantiate the class. This will extract terms from the the text_var column, embed those terms and embed the documents.
LysoMeta_MeSH_bunka = Bunka(data = LysoMeta_MeSH, #
                text_var = 'DescriptorName', #
                index_var = 'PMID',
                extract_terms=True,
                terms_embedding=True,
                docs_embedding=True,
                sample_size_terms=5000, #
                terms_limit=5000, #
                terms_ents=False,
                terms_ngrams=(2, 2),
                terms_ncs=False,
                terms_include_pos=["NOUN", "PROPN", "ADJ"],
                terms_include_types=["PERSON", "ORG"],
                terms_embedding_model="all-MiniLM-L6-v2",
                docs_embedding_model="all-MiniLM-L6-v2",
                language="en",
                terms_path=None,
                docs_dimension_reduction = 5,
                terms_embeddings_path=None,
                docs_embeddings_path=None,
                docs_multiprocessing = False,
                terms_multiprocessing = False)

# Extract the computed objects
LysoMeta_MeSH_terms = LysoMeta_MeSH_bunka.terms
LysoMeta_MeSH_terms_embeddings = LysoMeta_MeSH_bunka.terms_embeddings
LysoMeta_MeSH_docs_embeddings = LysoMeta_MeSH_bunka.docs_embeddings
```

```python
#The last argument label_mode was not present in original bunkatech package
#I added that in bunkatech/bunkatech/networks/networks_class.py
#line 375, line 98, line 128
#and altered code at line 490

#but I found another function (not method), so it's like 
#draw_network(LysoMeta_bunka) instead of LysoMeta_bunka.draw_network() )
#No change was made for that function
#That function is in bunkatech/bunkatech/networks/networks_graph.py

LysoMeta_MeSH_fig_2 = LysoMeta_MeSH_bunka.fit_draw(
            variables=["main form"],
            top_n=500,
            global_filter=0.2,
            n_neighbours=6,
            method="node2vec",
            n_cluster=10,
            bin_number=30,
            black_hole_force=3,
            color="community",
            size="size",
            symbol="entity",
            textfont_size=1,
            edge_size=1,
            height=1000,
            width=1000,
            template="none", #
            label_mode="top" #
        )
LysoMeta_MeSH_fig_2.show()
```

```python
import plotly
plotly.offline.plot(LysoMeta_MeSH_fig_2, filename = 'Z:\\YiboZhao\\NLP\\NLP_graph\\LysoMeta_MeSH_fig_2.html', auto_open=True)
```
